```python
import pandas as pd
import numpy as np
data={"name":["alice","bob","charlie","david"],
      "age": [ 25,np.nan,30,np.nan],
      "salary":[50000,60000,None,45000]}
df=pd.DataFrame(data)
print("before cleaning;\n",df)
```

```python
df_drop=df.dropna()
df["age"].fillna(df['age'].mean(),inplace=True)
df["salary"].fillna(df["salary"].median(),inplace=True)
print("\nafter cleaning:\n",df)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data=[10,12,13,12,11,200,14,13,15,12]
sns.boxplot(data)
plt.show()
Q1,Q3=np.percentile(data,[25,75])
IQR=Q3-Q1
lower,upper=Q1-1.5*IQR,Q3+1.5*IQR
outliers=[x for x in data if x<lower or x > upper ]
print("outliers detected:",outliers)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd
df=pd.DataFrame({"age":[18,22,30,40],
                 "salary":[20000,50000,80000,120000]})
scalar=MinMaxScaler()
df_norm=scalar.fit_transform(df)
scalar_std=StandardScaler()
df_std=scalar_std   .fit_transform(df)
print("original data:\n",df)
print("\nNormalized:\n",df_norm)
print("\nStandardized:\n",df_std)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
df=pd.DataFrame({
    "Name":["alice","bob","charlie"],
    "age":[25,32,47],
    "country":["US","India","US"],
    "spend1":[200,150,300],
    "spend2":[100,200,400],
    })
df["total_spend"]=df["spend1"]+df["spend2"]
df["age_group"]=pd.cut(df["age"],bins=[0,30,50],labels=["young","middle"])
encoder=LabelEncoder()
df["country_encoded"]=encoder.fit_transform(df["country"])
print(df)
```
